```python
import random

# Grid: 0 = empty, 1 = obstacle, T = task
grid = [
    ["0","0","T","0"],
    ["0","1","0","0"],
    ["0","0","0","T"],
    ["T","0","1","0"]
]

agent_pos = [0,0]

def print_grid():
    for i,row in enumerate(grid):
        line=""
        for j,cell in enumerate(row):
            if [i,j]==agent_pos: line+="A "   # Agent position
            else: line+=cell+" "
        print(line)
    print()

def get_percept(pos):
    i,j=pos
    return grid[i][j]

def reflex_agent():
    percept=get_percept(agent_pos)
    if percept=="T":
        print("Task found! Performing task...")
        grid[agent_pos[0]][agent_pos[1]]="0"
    else:
        # Move randomly but avoid obstacles
        moves=[(-1,0),(1,0),(0,-1),(0,1)]
        random.shuffle(moves)
        for dx,dy in moves:
            ni,nj=agent_pos[0]+dx,agent_pos[1]+dy
            if 0<=ni<4 and 0<=nj<4 and grid[ni][nj]!="1":
                agent_pos[0],agent_pos[1]=ni,nj
                break

# Run simulation
for step in range(10):
    print(f"Step {step+1}")
    print_grid()
    reflex_agent()
```
