```python
import time
from collections import deque

# Maze dimensions
W, H = 6, 6

# Maze grid
grid = [
    ['@', '.', '.', '#', '.', '.'],
    ['#', '#', '.', '#', '.', '#'],
    ['.', '.', '.', '.', '.', '.'],
    ['.', '#', '#', '#', '#', '.'],
    ['.', '.', '.', '.', 'E', '.'],
    ['#', '#', '#', '.', '#', '#']
]

# Agent start position
agent = (0, 0)

# Function to print the grid with the agent position and step number
def print_grid(agent_pos, step):
    print(f"Step {step}:")
    for i in range(W):
        row = ""
        for j in range(H):
            if (i, j) == agent_pos:
                row += 'A '
            else:
                row += grid[i][j] + ' '
        print(row)
    print("\n")

# BFS function to find shortest path to exit
def bfs(start):
    q = deque([(start, [start])])
    visited = {start}

    while q:
        (x, y), path = q.popleft()
        if grid[x][y] == 'E':
            return path
        for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < W and 0 <= ny < H and grid[nx][ny] != '#' and (nx, ny) not in visited:
                q.append(((nx, ny), path + [(nx, ny)]))
                visited.add((nx, ny))
    return None

# Find the escape path
path = bfs(agent)

# Animate the agent moving through the maze with step numbers
if path:
    print("Maze Escape Path:\n")
    step = 0
    for pos in path:
        step += 1
        print_grid(pos, step)
        time.sleep(1)
    print(f"Escape reached at {path[-1]} in Step {step}")
else:
    print("No path to the exit found.")
```

```python
from collections import deque
import time

# Grid dimensions
W, H = 5, 5

# Treasures, obstacles, agent start
treasures = [(0, 4), (2, 2), (4, 0)]
obstacles = {(1, 1), (3, 3)}
agent = (0, 0)
collected = []

# Function to print the grid
def print_grid(agent_pos, treasures, collected, step):
    print(f"Step {step}:")
    for i in range(W):
        row = ""
        for j in range(H):
            if (i, j) == agent_pos:
                row += 'A '
            elif (i, j) in obstacles:
                row += '# '
            elif (i, j) in treasures:
                row += 'T '
            elif (i, j) in collected:
                row += 'C '
            else:
                row += '. '
        print(row)
    print("\n")
    time.sleep(0.5)

# BFS function to find shortest path
def bfs(start, goal):
    q = deque([(start, [start])])
    visited = {start}
    while q:
        (x, y), path = q.popleft()
        if (x, y) == goal:
            return path
        for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:
            nx, ny = x+dx, y+dy
            if 0 <= nx < W and 0 <= ny < H and (nx, ny) not in obstacles and (nx, ny) not in visited:
                q.append(((nx, ny), path + [(nx, ny)]))
                visited.add((nx, ny))
    return None

# Initial display
step = 0
print("Initial Board:")
print_grid(agent, treasures, collected, step)
time.sleep(1)

# Move agent to collect treasures
for treasure in treasures:
    path = bfs(agent, treasure)
    if path:
        for pos in path[1:]:
            step += 1
            agent = pos
            if agent == treasure:
                collected.append(treasure)
            print_grid(agent, [t for t in treasures if t not in collected], collected, step)
    else:
        print(f"No path to treasure {treasure}")

print("Collected Treasures:", collected)
```

```python
sample_space = [1, 2, 3, 4, 5, 6]
A = [2, 4, 6]   # Even numbers
B = [5, 6]      # Numbers greater than 4

# Union and intersection using Python sets
union_AB = set(A) | set(B)          # A  B
intersection_AB = set(A) & set(B)   # A  B

# Probabilities
P_A = len(A) / len(sample_space)
P_B = len(B) / len(sample_space)
P_union = len(union_AB) / len(sample_space)
P_intersection = len(intersection_AB) / len(sample_space)

# Display
print("A =", A)
print("B =", B)
print("A  B =", union_AB)
print("A  B =", intersection_AB)
print(f"P(A) = {P_A:.2f}")
print(f"P(B) = {P_B:.2f}")
print(f"P(A  B) = {P_union:.2f}")
print(f"P(A  B) = {P_intersection:.2f}")
```
